#### Что такое return, почему он работает только с функциями?

Функции берут какие-то переменные производят какие-то операции и **возвращают** тоже какие-то переменные. **return** осуществляет это **возвращение** переменных. Таким образом **return** завершает выполнение функции.

**print()** только печатает, поэтому он может встречаться в теле функции несколько раз. А return -- не может.

Два важных примера:

```python
"""
в такой структуре функции принт и ретёрн взаимозаменяемы, потому что выполнится либо if, либо else,
то есть что-то напечатается или вернётся только один раз.
"""
def maximum_1(p, n):
    if p > n:
        print(p)
    else:
        print(n)
        
def maximum_2(p, n):
    if p > n:
        return p
    else:
        return n
    
maximum_1(1, 2)

maximum_2(3, 2)
```

```python
"""
Если мы хотим напечатать что-то еще, то использовать уже можно только принт
И снова есть два способа написать эту функцию
"""
def maximum_1(p, n):
    print("Так так, что тут у нас....")
    print("..........................")
    if p > n:
        print(p)
    else:
        print(n)
        
def maximum_2(p, n):
    print("Так так, что тут у нас....")
    print("..........................")
    if p > n:
        return p
    else:
        return n
    
maximum_1(1, 2)

maximum_2(3, 2)
```
